Now we will determine which features we would like to study. To reiterate, we want to find a correlation between age of diagnosis and presence of complications

#### Determine relevant features

```python
features = ['DIFAGE2', 'DIBAGE1',
    'HYPEV', 'HYPMDEV2', 'CHLEV', 'CHDEV', 'ANGEV', 'MIEV', 'HRTEV', 'STREV', 
    'EPHEV', 'COPDEV', 'ASPMEDEV', 'AASMEV', 'AASSTILL', 'AASMYR', 'AASERYR1', 
    'ULCEV', 'ULCCOLEV', 'CANEV', 'SINYR', 'CBRCHYR', 'KIDWKYR', 'LIVYR', 'ARTH1', 
    'ABLIND', 'VIM_GLEV', 'FLA1AR', 'DIBGDM', 'DIBINS2', 'DIBINS3', 
    'VIGLNGNO', 'VIGLNGTP'
]

type1_df2 = type1_df[features]
type1_df2.shape
```

```python
type1_df2.isnull().sum()
```

Too many missing variables. Need to drop more features. Need to also learn which features have "refused, don't know" responses.

```python
features_2 = []

for feature in features:
    if type1_df2[feature].isnull().sum() < 100:
        features_2.append(feature)
```

```python
type1_df3 = type1_df2[features_2]
type1_df3.isnull().sum()
```

```python
type1_df3
```

```python
bins = [0, 10, 20, 30, 40, 100]
labels = ['<10', '10-20', '20-30', '30-40', '40+']
type1_df3['Age_Group'] = pd.cut(type1_df3['DIBAGE1'], bins=bins, labels=labels)

type1_df3
```

```python
complications = ['HYPEV', 'CHLEV', 'CHDEV', 'ANGEV', 'MIEV', 'HRTEV', 'STREV', 'EPHEV',
                 'COPDEV', 'AASMEV', 'ULCEV', 'ULCCOLEV', 'CANEV', 'SINYR', 'CBRCHYR', 
                 'KIDWKYR', 'LIVYR', 'ARTH1', 'VIM_GLEV', 'FLA1AR']

# Replace 2 (No) with 0 and keep 1 (Yes) as 1

type1_df3[complications] = type1_df3[complications].replace({1: 1, 2: 0})

```

```python
type1_df3
```

```python
# Calculate mean complication rate by Age Group (as percentages)
complication_rates = type1_df3.groupby('Age_Group')[complications].mean() 
complication_rates
```

```python
plt.figure(figsize=(12, 8))
sns.heatmap(complication_rates, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Complication Rates by Age Group')
plt.ylabel('Complication')
plt.xlabel('Diagnosis Age Group')
plt.show()
```

```python
heart_complications = ['CHDEV', 'ANGEV', 'MIEV', 'HRTEV', 'STREV', 'CANEV']

for comp in heart_complications:
    sns.barplot(x='Age_Group', y=comp, data=type1_df3)
    plt.title(f'Prevalence of {comp} by Age Group')
    plt.ylabel('Proportion of Patients')
    plt.show()
```

```python
from scipy.stats import chi2_contingency

for comp in complications:
    contingency_table = pd.crosstab(type1_df3['Age_Group'], type1_df3[comp])
    chi2, p, dof, ex = chi2_contingency(contingency_table)
    print(f'{comp}: Chi-square test p-value = {p}')
```
